```python
import prince
import matplotlib
import pandas as pd
```

```python
col_reg = pd.read_csv('col_reg.csv')
```

```python
ca = prince.CA( ##set up the ca model
    n_components=2,
    n_iter=3,
    copy=True,
    check_input=True,
    engine='auto',
    random_state=42)
```

```python
ca = ca.fit(col_reg) #run the ca model
```

```python
ca.explained_inertia_ #get the inertias
```

```python
matplotlib.use('Agg')
ax = ca.plot_coordinates(
    X=col_reg,
    ax=None,
    figsize=(6, 6),
    x_component=0,
    y_component=1,
    show_row_labels=True,
    show_col_labels=True
)
```
