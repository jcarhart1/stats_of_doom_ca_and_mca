```python
import prince
import matplotlib
import pandas as pd
```

```python
chairs = pd.read_csv('chairs.csv')
```

```python
mca = prince.MCA( ##set up the mca analysis
    n_components=2,
    n_iter=3,
    copy=True,
    check_input=True,
    engine='auto',
    random_state=42)
```

```python
chairs = chairs.drop(['Shop', 'WordDE', 'Category'], axis=1)
```

```python
mca = mca.fit(chairs)
```

```python
mca.explained_inertia_
```

```python
ax = mca.plot_coordinates(
    X=chairs,
    ax=None,
    figsize=(6, 6),
    show_row_points=True,
    row_points_size=10,
    show_row_labels=False,
    show_column_points=True,
    column_points_size=30,
    show_column_labels=False,
    legend_n_cols=1
)
```
